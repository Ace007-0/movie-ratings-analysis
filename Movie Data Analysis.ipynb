```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('movie_metadata.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
rating_desc = df['imdb_score'].describe()
edges = [rating_desc['min'], rating_desc['25%'], rating_desc['50%'], rating_desc['75%'], rating_desc['max']]
labels = ['Poor', 'Average', 'Good', 'Excellent']

df['rating_category'] = pd.cut(df['imdb_score'], bins=edges, labels=labels, duplicates='drop')
df[['movie_title', 'imdb_score', 'rating_category']].head()
```

```python
print(df.columns)
```

* Movie Counts per Genre

```python
df = df.dropna(subset=['genres'])
df['genres_split'] = df['genres'].str.split('|')
df_exploded = df.explode('genres_split')

genre_counts = df_exploded['genres_split'].value_counts()

plt.figure(figsize=(12,6))
genre_counts.plot(kind='bar',color='skyblue')
plt.title('Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

* Average Rating per Genre

```python
genre_rating = df.groupby('genres')['imdb_score'].mean().sort_values(ascending=False)
genre_rating.plot(kind='bar', title='Average Rating per Genre', color='orange', figsize=(10, 5))
plt.ylabel('Average Rating')
plt.show()

# Below is same graph but after making difference in genres, and for that see genres_split.
```

```python
df = df.dropna(subset=['genres','imdb_score'])

df_exploded = df.explode('genres_split')

genre_rating = df_exploded.groupby('genres_split')['imdb_score'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
genre_rating.plot(kind='bar', color='orange')
plt.title('Average IMDb Rating per Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

* The 'genres' column contains multiple genres in a single string separated by '|', such as "Action|Drama".
* By using df['genres_split'] = df['genres'].str.split('|'), we convert each genre string into a list of individual genres.
* This allows us to analyze each genre separately. To enable this, we use df.explode('genres_split'), which creates a new row for each genre in the list, effectively duplicating the movie entry for each associated genre.
* For example, a movie with genres "Action|Drama" will now appear once under "Action" and once under "Drama".
* This approach ensures that each genre is counted accurately in aggregations like genre frequency or average rating per genre.


* Year-wise Rating Trend

```python
df.groupby('title_year')['imdb_score'].mean().plot(title='Average Movie Rating Over Years', figsize=(12,5))
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```
